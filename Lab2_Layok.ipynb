## Задание 5
1. Определите модуль разницы между средними тратами женщин и мужчин (трата - отрицательное значение amount).
2. Определите модуль разницы между средними поступлениями у мужчин и женщин

Обратите внимание, что для вычисления модуля разности точных знаний о том, 
какой класc относится к мужчинам, а какой - к женщинам, не требуется.

```python
women_spending = transactions[(transactions.amount < 0) & (transactions.gender == 0)].amount.mean()
men_spending = transactions[(transactions.amount < 0) & (transactions.gender == 1)].amount.mean()
module_spending = abs(women_spending - men_spending)

women_receipt = transactions[(transactions.amount > 0) & (transactions.gender == 0)].amount.mean()
men_receipt = transactions[(transactions.amount > 0) & (transactions.gender == 1)].amount.mean()
module_receipt = abs(women_receipt - men_receipt)

print("Модуль разницы между средними тратами женщин и мужчин: {}".format(module_spending))
print("Модуль разницы между средними поступлениями у мужчин и женщин: {}".format(module_receipt))
```

## Задание 8

1. Из поля tr_datetime выделите час tr_hour, в который произошла транзакция, как первые 2 цифры до ":". (\**)
2. Посчитайте ночные поступления для мужчин и женщин. Ночные - это примерно 00-06 часов. 

```python
transactions_copy = transactions.copy()
transactions_copy["tr_hour"] = transactions_copy.apply(lambda item: item.tr_datetime.split(":")[0].split()[1], axis=1).astype(int)
transactions_copy.head()

hour = transactions_copy.tr_hour
amount = transactions_copy.amount
gender = transactions_copy.gender

night_both_values = transactions_copy[(hour >= 0) & (hour <= 6) & (amount > 0)]

night_both_values.groupby(gender).amount.sum()
```
